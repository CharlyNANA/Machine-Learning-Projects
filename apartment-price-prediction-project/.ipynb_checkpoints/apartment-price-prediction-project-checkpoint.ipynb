## Preparation des données pour les algorithmes de machine learning

### Gestions Données manquantes et/ou constantes

+ Pour simplifier l'étude, nous avions supprimé les variables (colonnes) avec données manquantes.
    
+ Comme mentionné dans la section précédente, toutes les données constantes (i.e. de variance nul) seront simplement supprimées, car n'apportent pas d'informations pour la modélisation du prix de vente de l'immobilier dans la ville d'Ames.

### Gestion des variables qualitatives

Les algorithmes de Machine Learning ayant besoins des prédicteurs de types numériques, les variables qualitatives ont toutes été recodées suivant le **Codage disjonctif complet**. Cet opération transforme chaque modalité des variables qualitatives en variable binaire prenant la valeur 1 ou 0, selon que la modalité a été observée ou pas.

### Calibrage des données

Les données numériques ayant des échelles différentes ont été normalisé pour réduire le poids ou l'importance des grandes variables.

```python
###### Preparation des données pour les algorithmes de machine learning #######

# Supprime toutes les colonnes ayant au mois une données manquante
data = data.dropna(axis = 1, how = 'any')

# Transformation de la variable cible SalePrice en log
y_train = np.log(data.loc[ :,'SalePrice'])
data = data.drop(['SalePrice'], axis=1)

# Classe qui supprime la variable Id si elle existe dans le jeux de données initiale. Cette classe est concue
# dans le but de l'utiliser dans le pipeline comme estimateur 
from sklearn.base import BaseEstimator, TransformerMixin
class DeleteID(BaseEstimator, TransformerMixin):
    def __init__(self, delete_Id=True):
        self.delete_Id = delete_Id
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        if self.delete_Id:
            if ('Id' in list(X)):
                X = X.drop(['Id'], axis=1)
                return X
            else:
                return X
        else:
            return X

### Construction d'un pipeline pour gerer automatiquement les données numeriques et qualitatives
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.feature_selection import VarianceThreshold
from sklearn.preprocessing import OneHotEncoder, StandardScaler

# Pipeline pour pré-traité les données numeriques
num_pipeline = Pipeline([
    ('del', DeleteID(delete_Id=False)),
    ('var_thresold', VarianceThreshold(threshold=1e-1)),
    ('num_std_scaler', StandardScaler()),
])

# Pipeline pour pré-traité les données categorielles
cat_pipeline = Pipeline([
    ('one_hot_enc', OneHotEncoder()),
    ('cat_var_thresold', VarianceThreshold(threshold=1e-5)),
])

# On recupere la liste des attributs numeriques et catégorielles
cat_attributs = list(data.loc[:, data.dtypes == np.object])
num_attributs = list(data.loc[:, data.dtypes == np.int64])

# Pipeline final qui renvoie les données completes pré-traités pour les algorithme de machine Learning
full_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_attributs),
    ("cat", cat_pipeline, cat_attributs)], verbose = True)  

# Fonction qui renvoie la moyenne et la variance d'un vecteur
def MeanScore_and_StdScoce(vecteur_score):
    # print('RMSE Per Kfolds :', vecteur_score)
    print('Means Of RMSE : {:.5f}'.format(vecteur_score.mean()))
    print('Std Of RMSE : {:.5f}'.format(vecteur_score.std()))
    
    
```

```python
# Appel du pipeline concu pour le pré-traitement complet des données d'apprentissages
X_train = full_pipeline.fit_transform(data)
```

## Entrainement des modeles et evaluations

Dans cette partie, nous présentons les principaux résultats après entrainement des modèles. Pour chaque modèle entrainé, nous donnons, après avoir trouvés les meilleurs hyper paramètres sont **RSME** et sa **Variance** obtenue à partir des Kfols de la validation croisée.

### Regression Lineaire

```python
## LINEAR REGRESSION

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score

lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

# Estimation Erreur de prédiction : RMSE
lin_score = cross_val_score(lin_reg, X_train, y_train, scoring='neg_mean_squared_error', cv=100)
lin_score = np.sqrt(-lin_score)
MeanScore_and_StdScoce(lin_score)
```

### Nearest Neighbors Regression

```python
## Nearest Neighbors Regression

from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
# Recherche des meilleurs hyperparametres par quadrillage de la grille
neighbors_param_grid = {'n_neighbors' : np.arange(1, 10, 1), 
                        'weights' : ['uniform', 'distance'],
                        'algorithm' : ['auto', 'ball_tree', 'kd_tree'],
                        'metric' : ['euclidean', 'manhattan', 'chebyshev', 'minkowski']}


from sklearn.neighbors import KNeighborsRegressor
# Recherche aléatoire des meilleurs hyperparametres
neighbors_random = RandomizedSearchCV(estimator=KNeighborsRegressor(), param_distributions=neighbors_param_grid, 
                                      cv=20, random_state=42,n_iter=216, verbose=False, n_jobs=-1)

# Ajustement de la grille de recherche aux données
neighbors_random.fit(X_train, y_train)

# Estimation Erreur de prédiction : RMSE
neighbors_score = cross_val_score(neighbors_random.best_estimator_, X_train, y_train, 
                                  scoring='neg_mean_squared_error', cv=100)
# Affichage du RMSE 
neighbors_score =np.sqrt(-neighbors_score)
MeanScore_and_StdScoce(neighbors_score)
```

### Regression Ridge

```python
## RIDGE REGRESSION

# Recherche des meilleurs hyperparametres par quadrillage de la grille
ridge_param_grid = {'alpha' : np.arange(0, 1, 0.01), 
                    'solver' : ['auto', 'svd', 'cholesky', 'lsqr', 'sparse_cg', 'sag', 'saga'],
                    'fit_intercept' : ['True', 'False'],
                    'normalize' : ['True', 'Flase']}

# Creation d'une instance de RandomizedSearchCV
from sklearn.linear_model import Ridge
ridge_random = RandomizedSearchCV(estimator=Ridge(), param_distributions=ridge_param_grid, cv=20, 
                                  random_state=42,n_iter=1000, verbose=False, n_jobs=-1)

# Ajustement de la grille de recherche aux données
ridge_random.fit(X_train, y_train)

# Estimation Erreur de prédiction : RMSE
ridge_score = cross_val_score(ridge_random.best_estimator_, X_train, y_train, scoring='neg_mean_squared_error', cv=100)
ridge_score =np.sqrt(-ridge_score)
MeanScore_and_StdScoce(ridge_score)
```

### Regression Lasso

```python
## LASSO REGRESSION

from sklearn.linear_model import Lasso
# Recherche des meilleurs hyperparametres
lasso_param_grid = {'alpha' : np.arange(1e-6, 0.5, 1e-3), 
                    'fit_intercept' : ['True', 'False'],
                    'normalize' : ['True', 'Flase']}

# Creation d'une instance de RandomizedSearchCV
from sklearn.linear_model import Lasso
lasso_random = RandomizedSearchCV(estimator=Lasso(max_iter=9000), param_distributions=lasso_param_grid, cv=20, 
                                  random_state=42, n_iter=1000, verbose=False, n_jobs=-1)

# Ajustement de la grille de recherche aux données
lasso_random.fit(X_train, y_train)

# Estimation Erreur de prédiction : RMSE
lasso_score = cross_val_score(lasso_random.best_estimator_, X_train, y_train, scoring='neg_mean_squared_error', cv=100)
lasso_score =np.sqrt(-lasso_score)
MeanScore_and_StdScoce(lasso_score)
```

### Elastic Net

```python
## ELASTIC NET

# Definition d'une grille contenant une plage des hyperparametres
elastic_param_grid = {'alpha' : np.arange(0, 0.1, 1e-4), 
                      'l1_ratio' : np.arange(0.4, 0.6, 1e-3),
                      'fit_intercept' : ['True', 'False']}

# Creation d'une instance de RandomizedSearchCV
from sklearn.linear_model import ElasticNet
elastic_random = RandomizedSearchCV(estimator=ElasticNet(max_iter=5000), param_distributions=elastic_param_grid, 
                                    cv=20, random_state=42,n_iter=1000, verbose=False, n_jobs=-1)

# Ajustement de la grille de recherche aux données
elastic_random.fit(X_train, y_train)

# Estimation Erreur de prédiction : RMSE
elastic_score = cross_val_score(elastic_random.best_estimator_, X_train, y_train, scoring='neg_mean_squared_error', cv=100)
elastic_score =np.sqrt(-elastic_score)
MeanScore_and_StdScoce(elastic_score)
```

### Linear SVM Regression

```python
## LINEAR SVM REGRESSION 

# Definition d'une grille contenant une plage des hyperparametres importants
linear_svr_param_grid = {'epsilon' : np.arange(0, 0.5, 1e-2), 
                         'C' : np.arange(0, 2, 0.1), # C = Paramètre de régularisation.La force de la régularisation est inversement proportionnelle à C
                         'fit_intercept' : ['True', 'False'],
                         'loss' : ['epsilon_insensitive', 'squared_epsilon_insensitive']}

# Creation d'une instance de RandomizedSearchCV
from sklearn.svm import LinearSVR 
linear_svr_random = RandomizedSearchCV(estimator=LinearSVR(max_iter=90000), param_distributions=linear_svr_param_grid, 
                                       cv=10, random_state=42, n_iter=500, verbose=False, n_jobs=-1)

# Ajustement de la grille de recherche aux données
linear_svr_random.fit(X_train, y_train)

# Estimation Erreur de prédiction : RMSE
svm_score = cross_val_score(linear_svr_random.best_estimator_, X_train, y_train.values.ravel(), 
                            scoring='neg_mean_squared_error', cv=50)
svm_score =np.sqrt(-svm_score)
MeanScore_and_StdScoce(svm_score)
```

### Random Forest Regressor

### Limites et perspective d'amélioration possible

Lors de la recherche par quadrillage des meilleurs hyper paramètres de nos modèles d'apprentissages statistiques, les plages de valeurs (pour certain continue) de tous les hyper paramètres importants qui permettent d'optimiser les algorithmes et de contrôler la régularité des modèles n'ont pas été explorés car très couteux en temps de calcul. Nous nous sommes contentés de faire une recherche randomisée (grace à la classe *RandomizerSearchCV* de la library *sklearn*) des meilleurs hyper paramètres après discrétisation du domaine d'études des hyper paramètres important pour chaque modèles. 

Apres ce travail, on pourrait améliorer d'avantage la performance de nos modèles en faisant une recherche plus fines ou exhaustives (grâce à la classe *GridSearchCV* de la library *sklearn*) autours des valeurs de ces hyper paramètres trouvés précédemment (avec la classe *RandomizerSearchCV*), afin d'obtenir de meilleurs hyper paramètres.
